# Queston 4: Job Satisfaction

```python
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

import matplotlib.pyplot as plt
from collections import defaultdict
from IPython import display
%matplotlib inline

df = pd.read_csv('survey-results-public.csv')
schema = pd.read_csv('survey-results-schema.csv')
df.head()
```

## Finding proportion of missing values in Job satisfaction

```python
np.sum(df['JobSatisfaction'].isnull())/df.shape[0]
```

## Finding the group with highest job satisfaction as per Employment

```python
pd.DataFrame(df.query("JobSatisfaction != -1").groupby(['EmploymentStatus']).mean()['JobSatisfaction'])
```

## Analyzing whether large or small companies have higher job satisfaction

```python
company_size = pd.DataFrame(df.query("JobSatisfaction != -1").groupby(['CompanySize']).mean()['JobSatisfaction'])
company_size.sort_values(by=['JobSatisfaction'])
```

## Finding whether individuals who do programming as a hobby have higher job satisfaction or not

```python
hobby = pd.DataFrame(df.query("JobSatisfaction != -1").groupby(['ProgramHobby']).mean()['JobSatisfaction'])
hobby.sort_values(by=['JobSatisfaction'])
```

## Exploring whether people who work remotely have higher job satisfaction of not

```python
work_remote = pd.DataFrame(df.query("JobSatisfaction != -1").groupby(['HomeRemote']).mean()['JobSatisfaction'])
work_remote.sort_values(by=['JobSatisfaction'])
```

## Analyzing job satisfaction of individuals with higher formal education

```python
education = pd.DataFrame(df.query("JobSatisfaction != -1").groupby(['FormalEducation']).mean()['JobSatisfaction'])
education.sort_values(by=['JobSatisfaction'])
```
